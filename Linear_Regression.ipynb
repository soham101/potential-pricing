```python
# cost function module (h(x)=m_cur*x+b_cur)
def cost_function(m_cur,b_cur,x_train,y_train,m):
    lmt=m
    cost=0.0
    error=0.0
#     print(m_cur," ",b_cur)
    for i in range(0,lmt):
        error += (y_train[i]-(x_train[i]*m_cur+b_cur))**2
    error/=(2.0*lmt)
    print(error)
    #return error
```

```python
# gradient descent module
def gradient_des(m_cur,b_cur,y_train,x_train,lr,m):
    n=m
    b_grad=0.0
    m_grad=0.0
    for i in range (0,m):
        b_grad+=(1/n)*((m_cur*x_train[i]+b_cur)-y_train[i])
        
        m_grad+=(1/n)*((m_cur*x_train[i]+b_cur)-y_train[i])*x_train[i]
        
        b_cur=b_cur-(lr*b_grad)
        
        m_cur=m_cur-(lr*m_grad)
        
        #lr/=(.1*(i+1))
#         print(m_cur," ",b_cur)
    return [m_cur, b_cur]
```

```python
# main steps
epochs=2000
m_cur=0
b_cur=0
lr=0.0001
for i in range (0,epochs):
    m_cur, b_cur = gradient_des(m_cur,b_cur,y_train,x_train,lr,m)
    cost_function(m_cur,b_cur,x_train,y_train,m)
print(m_cur," ",b_cur)
```
